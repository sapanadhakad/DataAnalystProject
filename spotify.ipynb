# Heatmap between the numeric Features

```python
corr_df = d1.corr(method = "pearson")
```

```python
corr_df
```

```python
plt.figure(figsize = (14,6))
h = sns.heatmap(corr_df,annot=True,fmt = '1g',vmin = 1,vmax = 1,cmap="inferno",linewidths =1,linecolor ="Black")
h.set_title("correlation Heatmap Between Variable")
h.set_xticklabels(h.get_xticklabels(),rotation = 90)
```
